```python
import pandas as pd
import yfinance as yf
import time
```

```python
stock_list = pd.read_csv('/home/steve/documents/finance/obtain/symbols_master_list.csv')
```

```python
new_master_list = []
for ticker in stock_list['Symbol']:
    try:
        stock = yf.Ticker(ticker)
        hist = stock.history(period='max', interval = '1d')
    
        if len(hist) > 0:
            print('ticker '+ticker+' added.')
            new_master_list.append(ticker)
        time.sleep(3)
    except:
        pass
```

```python
print(len(new_master_list))
```

```python
all_stocks = pd.DataFrame()
```

```python
all_stocks['ticker'] = new_master_list
```
