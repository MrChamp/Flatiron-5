```python
from bs4 import BeautifulSoup
text_location = "/home/steve/documents/flatIron/fIProject/SentimentAnalysis/edgarFilings/sec-edgar-filings/TSLA/10-K/0001564590-18-002956/filing-details.html"

file = open(text_location)

soup = BeautifulSoup(file, "html.parser")
text = soup.get_text()
# print(text)
```

```python
import nltk
from nltk.corpus import stopwords
from wordcloud import WordCloud, STOPWORDS
```

```python
#Create stopword list:
stopwords = set(STOPWORDS)

stopwords.update(['br','href','s','u'])

#Generate wordcloud
wordcloud = WordCloud(stopwords=stopwords).generate(text)

plt.figure(figsize=(10,8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
